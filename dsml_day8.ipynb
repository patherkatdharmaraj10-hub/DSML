**The break statement**

```python
# break immediately exits the loop.
students  = ["shyam", "hari", "gita"]

for student in students:
  if student == "hari":
    break
  print(student)
```

Continue Statement

```python
# continue statement comparison value is not printed
students  = ["shyam", "hari", "gita"]

for student in students:
  if student == "hari":
   continue
  print(student)
```

```python
# Print only numbers that are divisible by 3 from 1 to 20 using continue.
for number in range(1, 21):
  if number % 3 != 0:
    continue
  print(number)
```

```python
# create a program where a secret number is stored inside the program. User keeps
# entering numbers. if the number matches, print "correct guess!", if wrong, print 'Try again'

secret_number = 1
while True:
  user_input = int(input("Enter a number:"))
  if user_input == secret_number:
    print("correct guess!")

  else:
    print('Try again')

```

```python
# Simple ATM Machine

balance = 1000  # Initial balance

while True:
    print("\n===== ATM MENU =====")
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Exit")

    choice = input("Enter your choice (1-3): ")

    if choice == "1":
        print(f"Your current balance is: Rs. {balance}")

    elif choice == "2":
        amount = float(input("Enter amount to deposit: Rs. "))

        if amount > 0:
            balance += amount
            print(f"Rs. {amount} deposited successfully.")
            print(f"New balance: Rs. {balance}")
        else:
            print("Invalid amount!")

    elif choice == "3":
        print("Thank you for using our ATM!")
        break

    else:
        print("Invalid choice! Please enter 1, 2, or 3.")
```
