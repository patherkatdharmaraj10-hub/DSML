```python
# Write a function that takes three numbers as arguments and returns the largest number.
def largest_num(num1, num2, num3):
    if num1 > num2 and num1 > num3:
        return num1
    elif num2 > num1 and num2 > num3:
        return num2
    else:
        return num3

result = largest_num(5, 35, 9)
print("The largest number is:", result)
```

enumerate() is a built-in Python function that lets you loop through an iterable (like a list or tuple) while keeping track of the index.

```python
# enumerate
fruits = ["a", "b", "c", ]
for index, fruit in enumerate(fruits):
  print(index, fruit)
```

for separate lists.
You need to use zip() to combine the lists, and then enumerate() to get the index.

```python
names = ["ram", "hari", "gita"]
ages = [18, 20, 39]
genders = ["male", "female", "other"]
salaries = [1888, 2000, 50000]

for index, (name, age, gender, salary) in enumerate(zip(names, ages, genders, salaries)):
    print(index, name, age, gender, salary)
```

map() in Python

map() is a built-in Python function used to apply a function to every element of an iterable (such as a list, tuple, or string).

Instead of writing a for loop, you can use map() to transform all elements.

Syntax
map(function, iterable)

or with multiple iterables:

map(function, iterable1, iterable2, ...)
Parameters
function → The function to apply.
iterable → The sequence (list, tuple, string, etc.).
Returns

A map object (iterator), which can be converted to a list, tuple, or set

```python
def sqrt(x):
  return x**2
```

```python
number = [1, 2, 3, 4, 5]
map_object = map(sqrt, number)
```

reduce():
reduce() is a function from the functools module that reduces an iterable (like a list) to a single value by repeatedly applying a function to its elements.

Unlike map(), which returns a transformed iterable, reduce() combines all the elements into one result.

Syntax
from functools import reduce

reduce(function, iterable[, initializer])

```python
from functools import reduce
number = [1, 3, 5, 9, 10]
result = reduce(lambda x, y: x + y, number)
print(result)



```

```python
# find maximum number using reduce function

from functools import reduce
number = [4, 9, 2, 8, 1]
max = reduce(lambda x, y: x if x > y else y, number)
print(max)



```
