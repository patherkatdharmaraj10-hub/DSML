**Write a python program to take marks as input from the user and display the grade using if and elif statements.
marks 8o to 100-> Grade A, between 60 and 79->B, 40 and 59->C, marks < 40->D**

```python
marks = input("Enter the marks:")
marks = float(marks)

if marks >= 80 and marks <= 100:
    print("Grade A")

elif marks >= 60 and marks <= 79:
    print("Grade B")

elif marks >= 40 and marks <= 59:
    print("Grade C")

else:
    print("Grade D")

```
