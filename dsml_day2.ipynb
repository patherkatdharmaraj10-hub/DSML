 # Data core Structures
 **Python provides four main built-in data structures for storing and organizing data.**

**1.List []:**
- ordered and mutable collection.
- allow duplicate values.
- stores mixed data types(float, integers, string, boolen).

***Examples***

```python
fruits = ["Apple", "Banana", "Mango", "Orange"]
print("fruits")
fruits.append("Grapes")
print(fruits)
fruits[:-1]
```

```python
fruits[:-1]
```

# Tuples():

- Starts with ()
- ordered and immutable faster than lists
- Data can not be changed/ modified
- Allows duplicate values

```python
coordinates = (10, 20, 40)
print(coordinates)
print(coordinates[-2])
```

# Set{}

- unrodered collection
- no duplicate values and mutable
- supports set operations like union, intersection, and differences


***Examples***

```python
unique_numbers = {1, 2, 3, 2, 1, 5}
print(unique_numbers)
```

```python
std_name = {"ram", "hari", "kamal", "gita"}
std_name2 = {"Divya", "Anita", "Kamala", "Sita", "ram"}
print(std_name.union(std_name2))
print(std_name)
```

# Dictionary {}

- Stores key-values pairs
- keys must be unique
- fast lookup
- very fast for searching using keys.

***Example***

```python
subject = {
    "name": "Machine Learning",
    "subject_code": "ML001",
    "credit_hours": 3
}
```

```python
subject["name"]
```

```python
Students = {
    "name": ["ram", "shyam", "hari"],
    "age": [18, 19, 20],
    "class": ["Bsc", "School", "Master"]
}
```

```python
print(Students["class"])
```
