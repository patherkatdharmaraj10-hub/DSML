# Write a program using user defined function (UDF)that:
- takes two numbers from the user,
- takes an operator sign (+, -, *, /),
- performs the selected operation,
- Displays the result

```python
def calculator(num1, num2, operator):
  if operator == "+":
    print(num1 + num2)

  elif operator =="-":
    print(num1 - num2)

  elif operator == "*":
    print(num1 * num2)

  elif operator == "/":
    if num2 == 0:
      print("cannot divide by 0")
    print(num1 / num2)

  else:
    print("Please enter the valid operator sign:")


num1 = int(input("enter the first number:"))
num2 = int(input("enter the second number:"))
operator = input("enter the operator:")


# function call
calculator(num1, num2, operator)

```

Student Result Management System
Create a user-defined function called student_result() that:
1.Takes the names and marks of 5 students as input.
2.Stores student data in a dictionary.
3.Uses conditional statements to assign grades:
A: 80 and above
B: 60–79
C: 40–59
F: Below 40
4.Displays each student's name, marks, and grade.
5.Calculates and displays:
Highest marks
Lowest marks
Average marks

```python
# Student Result Management System

def student_result():
    students = {}

    # Input names and marks
    for i in range(5):
        name = input(f"Enter name of student {i+1}: ")
        marks = float(input(f"Enter marks of {name}: "))
        students[name] = marks

    print("\n------ Student Results ------")

    marks_list = []

    # Display results and assign grades
    for name, marks in students.items():

        if marks >= 80:
            grade = "A"
        elif marks >= 60:
            grade = "B"
        elif marks >= 40:
            grade = "C"
        else:
            grade = "F"

        print(f"Name: {name}")
        print(f"Marks: {marks}")
        print(f"Grade: {grade}")
        print("------------------------")

        marks_list.append(marks)

    # Calculate statistics
    highest = max(marks_list)
    lowest = min(marks_list)
    average = sum(marks_list) / len(marks_list)

    print("\n------ Statistics ------")
    print("Highest Marks:", highest)
    print("Lowest Marks :", lowest)
    print("Average Marks:", average)


# Function Call
student_result()
```

Smart Shopping Cart System
Create a user-defined function called shopping_cart() that:
1.Allows the user to enter 5 product prices.
2.Stores the prices in a list.
3.Uses a loop to calculate the total bill.
4.Applies discounts using conditional statements:
Total ≥ 5000 → 15% discount
Total ≥ 3000 → 10% discount
Total ≥ 1000 → 5% discount
Otherwise → No discount
5.Displays:
Product list
Total amount
Discount amount
Final payable amoun

```python
def shopping_cart():

    # Step 1: Create an empty list
    product_prices = []

    # Step 2: Accept prices of 5 products
    print("Enter the prices of 5 products:")

    for i in range(1, 6):
        price = float(input(f"Product {i} Price: Rs. "))
        product_prices.append(price)

    # Step 3: Calculate total bill
    total = 0

    for price in product_prices:
        total += price

    # Step 4: Apply discount
    if total >= 5000:
        discount_rate = 15
    elif total >= 3000:
        discount_rate = 10
    elif total >= 1000:
        discount_rate = 5
    else:
        discount_rate = 0

    discount_amount = total * discount_rate / 100
    final_amount = total - discount_amount

    # Step 5: Display bill
    print("\n" + "=" * 40)
    print("        SMART SHOPPING CART")
    print("=" * 40)

    print("\nProduct Prices")

    for i, price in enumerate(product_prices, start=1):
        print(f"Product {i}: Rs. {price:.2f}")

    print("\n" + "-" * 40)
    print(f"Total Amount      : Rs. {total:.2f}")
    print(f"Discount ({discount_rate}%) : Rs. {discount_amount:.2f}")
    print(f"Final Amount      : Rs. {final_amount:.2f}")
    print("=" * 40)


# Function Call
shopping_cart()
```
