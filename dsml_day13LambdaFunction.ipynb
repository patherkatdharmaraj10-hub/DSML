The reduce() function is a built-in function (available in the functools module) that reduces a sequence (like a list or tuple) to a single value.

It repeatedly applies a function to the elements of the sequence until only one result remains.

from functools import reduce

reduce(function, iterable)

```python
# using reduce(): sum of all numbrs -> lambda
from functools import reduce
numbers = [1, 2, 3, 4, 5, 6]
sum_numbers = (reduce(lambda x, y: x + y, numbers))
```

The filter() function is a built-in Python function used to select or keep only those elements that satisfy a given condition.

filter(function, iterable)

```python
marks = [ 80, 89, 90, 100, 99]

grade = filter(lambda x: x > 80, marks)
print(grade)
```

What is map()?

The map() function is a built-in Python function that applies the same operation to every element of an iterable (such as a list or tuple) and returns the transformed values.

Simple Definition

map() changes every element in a collection according to a function.

```python
name = ["ram", "hari", "kamal", "giat"]
Name = list(map(lambda x:x.upper(), name))

print(Name)


```

```python
name = "ram kumar sharma"
word = name.split()
print(word)

```

```python
len_name = map(lambda x:len(x), word)

print(list(len_name))
```

Interview Answer:
Use lambda for small, anonymous, one-line functions that are used temporarily, especially with functions like map(),
filter(), and sorting. Use def when the function contains multiple statements, complex logic, needs documentation, or will
be reused throughout the program.


What is zip()?

The zip() function combines two or more iterables (lists, tuples, strings, etc.) element by element into tuples.

```python
name = ["ram", "hari", "kamal", "gita"]
marks = [ 80, 89, 90, 100, 99]

result = zip(name, marks)
print(list(result))
```

List Comprehension with if-else

Syntax

[expression_if_true if condition else expression_if_false for item in iterable]

```python
# numbers = [1, 2, 3, 4, 5, 6]
# output
# ["odd", "even", "odd", "enven", "odd", "even"]

numbers = [1, 2, 3, 4, 5, 6]
["even" if num % 2 == 0 else "odd" for num in numbers]
```

```python
lst = [1, 2, 3, 4, 5, 6, 7, 8, 90]
result = ["max" if x > 80 else "min" for x in lst]
print(result)
```
