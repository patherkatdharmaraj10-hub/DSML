***Q.N.2. You are given a dictionary of a students and marks. Print the student with the highest marks.***

```python
# Student dictionary
students = {    
    "john": 25,
    "jane": 92,     
    "hari": 24,
    "gita": 91,
    "Sita": 91
}

top_student = max(students, key = students.get)
print("Top student name:", top_student)
print("Top student marks:", students[top_student])

```

### List Operations

```python
numbers = [1, 4, 3, 4, 5]
numbers.append(6)
print("After appending 6:", numbers)
```

```python
numbers1 = [10, 40, 30, 40, 50]
numbers.extend(numbers1)
print("After extending numbers with numbers1:", numbers)
```

```python
numbers.insert(2, 99) # index and valu("", "")
print("After inserting 99 at index 2:", numbers)
```

```python
# remove
numbers.remove(4) # remove first occurrence of 4
print("After removing 4:", numbers)
```

```python
# pop: removes and returns the last item if index is not specified, otherwise removes and returns the item at the specified index.
popped_item = numbers.pop()  # removes the last item
```

```python
numbers.clear()  # removes all items from the list
```

```python
# index 
n = [1, 20, 40, 60]
index1= n.index(40)  # find the index of the first occurrence of 4
```

```python
# count: counts the numb
# er of occurrences of a specified value in the list.
count1 = n.count(40)  # count the occurrences of 40 in the list
print("Count of 40 in the list:", count1)
```

```python
# sort: sorts the list in ascending order by default. You can also specify a custom sorting order using the key parameter.
n.sort()  # sorts the list in ascending order
print("After sorting:", n)
```

```python
# reverse: reverses the order of the list.
print("Before reversing:", n)
n.reverse()  # reverses the order of the list
print("after reversing:", n)
```

```python
# copy: returns a shallow copy of the list.
copied_list = n.copy()  # creates a shallow copy of the list
print("Copied list:", copied_list)
```

```python
import copy

deep_copied_list = copy.deepcopy(n)  # creates a deep copy of the list
original_list = [1, 2, [3, 4]]
shallow_copied_list = original_list.copy()
print("Original list:", original_list)
```

```python
# deep copy: creates a new list that is a copy of the original list, including all nested lists.
import copy
original_list = [1, 2, [3, 4]]  
deep_copied_list = copy.deepcopy(original_list)  # creates a deep copy of the list
print("Original list:", original_list)
print("Deep copied list:", deep_copied_list)
```

```python
# using Lists as Queues
from collections import deque # deque means Double Ended Queue, w   hich allows you to efficiently add and remove elements from both ends of the queue.
queue = deque(["apple", "banana", "cherry"])
queue.append("grapes") # adds "grapes" to the right end of the queue
queue.appendleft("kiwi")  # adds "kiwi" to the left end of the queue
print("Queue after removing leftmost item:", queue)
```

```python
queue.popleft()  # removes and returns the leftmost item from the queue
print("Queue after removing leftmost item:", queue)
```
