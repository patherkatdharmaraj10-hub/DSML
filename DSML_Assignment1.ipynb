***Q.N.1. Create a Python program using at least 10 string methods from the table above and show their outputs with examples. Try methods like upper(), split(), replace(), find(), and strip().***

```python
# String
text = " Hello, Welcome to Kathmandu, Nepal   "
print("Original String is: ")
print(text)
```

```python
# 1.upper()
print("After using upper() method:")
print(text.upper())
```

```python
# 2.lower()
print("After using lower() method:")
print(text.lower())
```

```python
# 3.title()
print("After using title() method:")
print(text.title())
```

```python
# 4.strip()
print("After using strip() method:")
print(text.strip())
```

```python
# 5.replace()
print("After using replace() method:")
print(text.replace("Kathmandu", "Pokhara"))
```

```python
# 6. split()
print("After using split() method:")
print(text.split())
```

```python
# 7.find()
print("After using find() method:")
print(text.find("hello"))
```

```python
# 8.count()
print("After using count() method:")
print(text.count("hello"))
print(text.count("l"))
```

```python
# 8.endswith()
print("After using endswith() method:")
print(text.endswith("Nepal"))
```

```python
# 8.startswith()
print("After using startswith() method:")
print(text.strip().startswith("Hello"))
```

***Q.N.2. Take a sentence from the user and perform differeent string operations on it such as converting case, coounting words, checking digits/alphabet, and replacing words. Print all results clearly with labels.***

```python
# user input
sentence = input("Enter a sentence.")
```

```python
print("original sentence:", sentence)
```

```python
# Counting words
print(sentence.count("l"))
```

```python
# converting case
print(sentence.capitalize())
```

```python
# count total characters
print("total characters  :", len(sentence))
```

```python
# count total words
word = sentence.split()
print("total words:",len(word))
```

```python
# check i the sentence contains only digits
print("is digit only:", sentence.isdigit())
```

```python
# check if the sentence contains only alphabets
print("is albhabet only?", sentence.isalpha())
```

```python
print("is alphabet only?", sentence.replace("","").isalpha())
```

```python
# replacing a word
old_word = input("\nEnter the old word:")
new_word = input("Enter the new word:")
print("updated word is:",new_word.replace("",""))
```
