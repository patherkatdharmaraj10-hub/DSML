What is a Docstring?

Docstring means:

Documentation String

It is a special string written inside a function, class, or module ( simply a python file
 like math, os, sys, function tool) to explain what it does.

A docstring is written using:

```python
"""
Triple Double Quotes
"""
```

Why do we use Docstrings?

Suppose after 6 months you open your project.

Without docstring:

def abc(x, y):
    return x*y

You may forget:

What is abc?
What are x and y?

4. Where is a Docstring Written?

It must be the first statement inside a function.

Accessing the document

```python
def message():
  """
  hello, how are you?
  """
print(message.__doc__)
```

```python
def message():
  """
  hello, how are you?
  """

help(message)
```

Explain functions/classes/modules

```python
def addition(num1, num2):
    """
    Function Name:
        addition

    Purpose:
        Adds two numbers.

    Parameters:
        num1 : First Number
        num2 : Second Number

    Returns:
        Sum of numbers
    """

    return num1 + num2


number1 = float(input("Enter first number: "))
number2 = float(input("Enter second number: "))

result = addition(number1, number2)

print("Sum =", result)

print("\nDocstring:\n")
print(addition.__doc__)
```

*args and **kwargs in Python


What is *args?

*args allows a function to accept any number of positional arguments.

Instead of receiving one value, it collects all extra positional arguments into a tuple.

3. Meaning of the Asterisk (*)

The * means:

Collect multiple positional arguments.

Positional vs Keyword Arguments
Positional Arguments	Keyword Arguments

Matched by order	Matched by parameter name

Order matters	Order doesn't matter

Example: student("Ram", 20)	Example: student(name="Ram", age=20)

```python
def full_name(*args):
  if len(args) ==2:
    full_name = args[0] + " " + args[1]
    return full_name

  else:
    return "invalid input"
full_name("Ram", "Sharma")
```

What is **kwargs?
The special syntax **kwargs allows us to pass any number of keyword
arguments (arguments in the form key=value). These arguments are
collected into a dictionary, where:
Keys = argument names
Values = argument values
def function_name(**kwargs):
statements


```python
def student(**kwargs):
    print(kwargs)
```

```python
student(name="Ram", age=20)
```

```python
# student detail
def student(**details):
  print(details)

student(name = "ram", age = "18", city = "kathmandu")
```

Write a function using **kwargs that prints the complete dictionary.

```python
def print_dict(**kwargs):
  print(kwargs)
  # return kwargs

print_dict(name = "ram", age = 18, add = "ktm")
```

Write a function using **kwargs that prints only the keys.

```python
def print_keys(**kwargs):
  for key in kwargs:
    print(key)


print_keys(name = "ram", age = 18, add = "ktm")
```

Write a function using **kwargs that prints only the values.

```python
def print_values(**kwargs):
  for value in kwargs.values():
    print(value)

print_values(name = "ram", age = 18, add = "ktm")
```

Write a function using **kwargs that prints each key-value pair.

```python
def print_key_value(**kwargs):
  for key, value in kwargs.items():
    print(key, value)

print(print_key_value(name = "ram", age = 18))
```

Write a function using **kwargs that checks whether "name" exists.

```python
def name_exists(**kwargs):
  if "name" in kwargs:
    print("name exists")
  else:
    print("name does not exist")

name_exists(name = "ram", age = 18)
```

```python
def sum(*numbers):
  total = 0

  for n in numbers:
    total += n
  return total
```

```python
def area(**kwargs):
  return kwargs["length"] * kwargs["width"]

area(length = 10, width = 20)
```

Write a function to calculate the perimeter of a rectangle.

```python
def perimeter(**kwargs):
  return  2 * (kwargs["length"] + kwargs["width"])

perimeter(length = 10, width = 20)
```

Create a function cart_total(*args) that accepts the prices of any number of products and returns the total bill amount.

```python
def cart_total(*args):
    total = 0

    for price in args:
        total += price

    return total


bill = cart_total(250, 500, 120, 80, 350)

print("Total Bill = Rs.", bill)
```

Create a funcion product_info(**kwargs) that accepts product details such as name, price, brand, and category, then displaysthem in  a readable format.

```python
def product_info(**kwargs):

    print("Product Information")
    print("-" * 30)

    for key, value in kwargs.items():
        print(f"{key.capitalize()} : {value}")


product_info(
    name="Laptop",
    price=85000,
    brand="Dell",
    category="Electronics"
)
```
